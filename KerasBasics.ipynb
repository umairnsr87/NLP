```python
import seaborn as sns
import matplotlib.pyplot as mlp
import pandas as pd
import numpy as np
import tensorflow as tf
import keras
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
X=iris.data
Y=iris.target
```

```python
#for multiclass classification
#we need to provide the target variable into one hot encoding method or matrix
from keras.utils import to_categorical
Y=to_categorical(Y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=42,test_size=.33)
```

```python
#for neural network it is the best way to scale and standardize the data
from sklearn.preprocessing import MinMaxScaler
```

```python
scalar_object=MinMaxScaler()
scalar_object.fit(X_train)
```

```python
X_train=scalar_object.transform(X_train)
#now we will scale the test data
X_test=scalar_object.transform(X_test)
```

```python
#we are importing the sequential models to the keras to compute 
from keras.models import Sequential
from keras.layers import  Dense
```

```python
#step 2 made th esequential model
model=Sequential()
#add the layers,
model.add(Dense(8,input_dim=4,activation='relu'))

model.add(Dense(8,input_dim=4,activation='relu'))
model.add(Dense(3,activation='softmax'))
model.compile(loss="categorical_crossentropy",optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train,Y_train,epochs=150,verbose=2)
```

```python
predicted=model.predict_classes(X_test)
```

```python
from sklearn.metrics import auc,classification_report,confusion_matrix
```

```python
#to change back Y_test into its original state (reversing one hot encoding)
y_test=Y_test.argmax(axis=1)
```

```python
y_test
```

```python
confusion_matrix(y_test,predicted)
```

```python
print(classification_report(y_test,predicted))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predicted)
```
