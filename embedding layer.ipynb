```python
import tensorflow
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
tf.__version__
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=['a glass of milk',"a glass of juice","the cup of tea","i am a good boy","i am a good developer",
     "understand the meaning of words","your videos are shit"]
```

```python
sent
```

```python
voc_size=10000
```

```python
one_hot_enc=[one_hot(sen,voc_size) for sen in sent]
one_hot_enc
```

```python
sent_length=8
embedded_docs=pad_sequences(one_hot_enc,padding="pre",maxlen=sent_length)
embedded_docs
```

```python
model=Sequential()
model.add(Embedding(voc_size,10,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python


print(model.predict(embedded_docs))

```
